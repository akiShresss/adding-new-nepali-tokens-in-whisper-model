```python
from transformers import AutoTokenizer

# Load the Whisper tokenizer
tokenizer = AutoTokenizer.from_pretrained("openai/whisper-small")

# Define a sample Nepali sentence
sample_sentence = "काठमाडौंमा पहिलो पटक स्ट्रबेरीको व्यवसायिक खेती सुरु भएको छ।"

# Function to tokenize each word in the sentence and print tokens
def tokenize_and_print(sentence, tokenizer):
    print("Whisper Tokenization:")
    total_tokens = 0

    # Tokenize each word in the sentence individually
    for word in sentence.split():  # Split by spaces to get individual words
        tokens = tokenizer(word).input_ids
        print(f"Word: {word} -> Tokens: {tokens}")
        total_tokens += len(tokens)

    print(f"Total number of tokens: {total_tokens}")

# Tokenize and print the result
tokenize_and_print(sample_sentence, tokenizer)
```

```python
import os
from datasets import load_dataset
from transformers import AutoTokenizer
import re

# Set up Hugging Face token
os.environ['HF_TOKEN'] = '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'

# Load the Nepali text dataset
dataset = load_dataset("Sakonii/nepalitext-language-model-dataset", split="train")

# Define a function to filter out Nepali words
def filter_nepali_text(text):
    # Nepali characters Unicode range
    nepali_pattern = re.compile(r'[\u0900-\u097F]+')
    return ' '.join(nepali_pattern.findall(text))

# Apply the filter to the dataset
filtered_texts = [filter_nepali_text(example["text"]) for example in dataset if filter_nepali_text(example["text"])]

# Check a sample of the filtered data
print("Sample filtered text:", filtered_texts[:5])

# Define a batch iterator for the filtered texts
batch_size = 1000
def batch_iterator():
    for i in range(0, len(filtered_texts), batch_size):
        yield filtered_texts[i : i + batch_size]

# Initialize and train the BPE tokenizer
tokenizer = AutoTokenizer.from_pretrained("openai/whisper-small")
new_tokenizer = tokenizer.train_new_from_iterator(batch_iterator(), vocab_size=25000)

# Define the directory to save the tokenizer
save_directory = "./nepali_bpe_tokenizer"
os.makedirs(save_directory, exist_ok=True)

# Save the trained tokenizer
new_tokenizer.save_pretrained(save_directory)

print(f"Tokenizer saved to {save_directory}")
```

```python
import os
from datasets import load_dataset
from transformers import AutoTokenizer
import re
from collections import Counter

# Set up Hugging Face token
os.environ['HF_TOKEN'] = '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'

# Define save directory for the extended tokenizer
extended_save_directory = "./extended_nepali_whisper_tokenizer"

# Check if the tokenizer directory already exists
if os.path.exists(extended_save_directory):
    print(f"Tokenizer already exists at {extended_save_directory}. Loading existing tokenizer.")
    tokenizer = AutoTokenizer.from_pretrained(extended_save_directory)
else:
    # Load the Nepali text dataset
    dataset = load_dataset("Sakonii/nepalitext-language-model-dataset", split="train")

    # Define a function to filter out Nepali words
    def filter_nepali_text(text):
        nepali_pattern = re.compile(r'[\u0900-\u097F]+')
        return ' '.join(nepali_pattern.findall(text))

    # Apply the filter to the dataset
    filtered_texts = [filter_nepali_text(example["text"]) for example in dataset if filter_nepali_text(example["text"])]

    # Collect unique words and their frequency
    word_counter = Counter(" ".join(filtered_texts).split())
    most_common_words = [word for word, _ in word_counter.most_common(1000)]  # Limit to top 1000 frequent words

    # Initialize the Whisper tokenizer
    tokenizer = AutoTokenizer.from_pretrained("openai/whisper-small")

    # Check existing tokens in Whisper's vocabulary
    existing_tokens = set(tokenizer.get_vocab().keys())

    # Identify new tokens that are not in the existing vocabulary
    new_tokens = set(most_common_words) - existing_tokens

    # Add new tokens to the tokenizer's vocabulary
    tokenizer.add_tokens(list(new_tokens))

    # Save the extended tokenizer
    os.makedirs(extended_save_directory, exist_ok=True)
    tokenizer.save_pretrained(extended_save_directory)
    print(f"Extended tokenizer saved to {extended_save_directory}")

# Define the sample sentence
sample_sentence = "काठमाडौंमा पहिलो पटक स्ट्रबेरीको व्यवसायिक खेती सुरु भएको छ।"

# Tokenize the entire sample sentence using the extended vocabulary
tokens = tokenizer.encode(sample_sentence, add_special_tokens=False)
total_tokens = len(tokens)

# Print each token with its corresponding subwords or characters
print("\nExtended Whisper Tokenization:")
for token_id in tokens:
    # Decode the token to see the actual subword or character
    subword = tokenizer.decode([token_id])
    print(f"Token ID: {token_id} -> Subword: {subword}")

# Print the total number of tokens
print(f"Total number of tokens: {total_tokens}")
```
